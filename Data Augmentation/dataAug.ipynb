```python
# example of horizontal shift image augmentation
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
# load the image
img = load_img('bird.jpg')
# convert to numpy array
data = img_to_array(img)
# expand dimension to one sample
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-200,200])
# prepare iterator
it = datagen.flow(samples, batch_size=1)
# generate samples and plot
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```

# **Create operation convert to RGB all the images**

```python
from Augmentor.Operations import Operation
from PIL import ImageFilter
import random
# Create your new operation by inheriting from the Operation superclass:
class ConvertoRGB(Operation):
    # Here you can accept as many custom parameters as required:
    def __init__(self, probability):
        # Call the superclass's constructor (meaning you must
        # supply a probability value):
        Operation.__init__(self, probability)
        # Set your custom operation's member variables here as required:
        # self.max_kernel = max_kernel

    # Your class must implement the perform_operation method:
    def perform_operation(self, images):
        # Start of code to perform custom image operation.
        #factor = round(random.uniform(self.min_factor, self.max_factor), 2)
        def do(image):
          return image.convert('RGB')
        augmented_images = []
        for image in images:
          augmented_images.append(do(image))
        return augmented_images
```

# **Create a new opertion (Blurry operation)**

```python
from Augmentor.Operations import Operation
from PIL import ImageFilter
import random
# Create your new operation by inheriting from the Operation superclass:
class GaussianBlur(Operation):
    # Here you can accept as many custom parameters as required:
    def __init__(self, probability):
        # Call the superclass's constructor (meaning you must
        # supply a probability value):
        Operation.__init__(self, probability)
        # Set your custom operation's member variables here as required:
        # self.max_kernel = max_kernel

    # Your class must implement the perform_operation method:
    def perform_operation(self, images):
        # Start of code to perform custom image operation.
        #factor = round(random.uniform(self.min_factor, self.max_factor), 2)
        def do(image):
          return image.filter(ImageFilter.GaussianBlur(radius=2))
        augmented_images = []
        for image in images:
          augmented_images.append(do(image))
        return augmented_images
```
